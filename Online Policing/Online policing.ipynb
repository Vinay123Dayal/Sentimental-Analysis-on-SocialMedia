```python
### Methods to find PII--> 10 digits number(phone number), 

import re

count=0;
for i in replylist:
  text=i.text;
  #Phone number matches
  #print(text);
  pat=re.compile(r'\d\d\d\d\d\d\d\d\d\d');
  match=re.findall(pat,text);
  if(match):
    count+=1;
    print(match);
print("Phone numbers as PII: ",count);
count=0;

## Reply of mail
pat2="([a-zA-Z0-9_\.-]+)@([\da-zA-Z\.-]+)\.([a-zA-Z\.]{2,6})"
for i in replylist:
  text=i.text;
  match=re.findall(pat2,text);
  if(match):
    print(match);
    count+=1;
print("Mail as PII: ",count);
## Aadhar Number Found
count=0;
part2="(^[2-9]{1}[0-9]{3}\\s[0-9]{4}\\s[0-9]{4}$)";
for i in replylist:
  text=i.text;
  match=re.findall(part2, text);
  if(match):
    print(match);
    count+=1;
print("Aadhar as PII: ",count);
# Vehicle Number found
count=0;
part2="([a-zA-Z]{2})(\d{2})([a-zA-Z]{2})(\d{4})";
for i in replylist:
  text=i.text;
  match=re.findall(part2, text);
  if(match):
    print(match);
    count+=1;
print("Vehicle number plate as PII: ",count);
```

```python
import datetime
import numpy as np


def finddifference(time1, time2):
  c=time1-time2;
  return c.seconds/60;

time=[];
for i in timelist:
  time.append(finddifference(i[0],i[1]));
minimum=np.min(time);maximum=np.max(time);mean=np.mean(time);standard_deviation=np.std(time);
print("Minimum: ",minimum);print("Maximum: ",maximum);print("Mean: ",mean);print("Standard Deviation: ",standard_deviation);
```

```python
##Making Time Series plot
import matplotlib.pyplot as plt
import plotly.graph_objects as go
dates1={};hours={};
for i in timelist:
  d1=i[1].strftime("%x");d2=i[1].strftime("%X");
  d1 = d1[-2:]+"/"+d1[:2]+"/"+d1[3:5]
  if(d1 not in dates1.keys()):
    dates1[d1]=0;
  if(d2 not in hours.keys()):
    hours[d2]=0;
  dates1[d1]+=1;
  hours[d2]+=1;
sorteddate=sorted(dates1.items(),key=lambda item: item[0])
l1=[];l2=[];
for i in sorteddate:
  l1.append(i[0]);l2.append(i[1]);
fig_id=go.Figure([go.Scatter(x=l1, y=l2)])
fig_id.show();
```

```python
from textblob import TextBlob
crimes = []
appreciation = []
request = []
for i in replylist:
  text=i.text;
  first = TextBlob(text).sentiment
  #print(blob1)
  if(first.polarity > 0.5):
    appreciation.append(i)
  elif (first.polarity < -0.5):
    crimes.append(i)
  elif(first.polarity > -0.2 and first.polarity < 0.2):
    request.append(i)
print("crimes labeled: ",len(crimes))
for i in crimes:
  print(i.text);
print("appreciation labeled: ",len(appreciation))
for i in appreciation:
  print(i.text);
print("Request: ",len(request))
```
